```python
import pandas as pd
df = pd.read_csv(r"C:\Users\naysa\OneDrive\Documents\ecs 171\Spam Emails\spambase.data", sep=',', header=None)
print(df.head())
```

```python

# Define column headers for 'Percentage of words that match WORD'
word_columns = [f'Percentage of words that match WORD{i}' for i in range(1, 49)]

# Define column headers for 'Percentage of words that match CHAR'
char_columns = [f'Percentage of words that match CHAR{i}' for i in range(1, 7)]

# Define the remaining column headers
remaining_columns = [
    'average length of uninterrupted sequences of capital letters',
    'length of longest uninterrupted sequence of capital letters',
    'total number of capital letters in the e-mail',
    'class'
]

# Concatenate all column headers
column_names = word_columns + char_columns + remaining_columns

# Assign the column names to the DataFrame
df.columns = column_names

# Now you have your data in a DataFrame 'df' with the specified headers
print(df.head())  # Display the first few rows of the DataFrame
```

```python
#count rows of missing data for each column
missing_number=df.isnull().sum().sort_values(ascending=False)
#Calculate percentage of null value
missing_percentage=missing_number/len(df)
missing_info=pd.concat([missing_number,missing_percentage],axis=1,keys=['missing number','missing percentage'])
missing_info.head(20)
```
